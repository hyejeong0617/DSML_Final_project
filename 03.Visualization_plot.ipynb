```python
import matplotlib.pyplot as plt
import numpy as np

# -------------------------------
# Data: Training vs Test
# -------------------------------

models = [
    "Best XGBoost",
    "Overfitted XGBoost",
    "Dirty Dataset XGBoost"
]

# Training scores: [accuracy, F1]
train_scores = np.array([
    [0.9332, 0.93],   # Best XGBoost
    [0.9737, 0.97],   # Overfitted
    [0.9203, 0.92]    # Dirty dataset
])

# Test scores: [accuracy, F1]
test_scores = np.array([
    [0.8736, 0.87],   # Best XGBoost
    [0.8775, 0.88],   # Overfitted
    [0.8658, 0.86]    # Dirty dataset
])

metrics = ["Accuracy", "F1 Score"]
x = np.arange(len(models))  # model positions
width = 0.35  # bar width

# -------------------------------
# Plot
# -------------------------------
plt.figure(figsize=(10,6))

# Training bars
plt.bar(x - width/2, train_scores[:,0], width, label="Train Accuracy", color='steelblue')
plt.bar(x - width/2, train_scores[:,1], width, bottom=train_scores[:,0]*0, label="Train F1", color='skyblue', alpha=0.7)

# Test bars
plt.bar(x + width/2, test_scores[:,0], width, label="Test Accuracy", color='darkorange')
plt.bar(x + width/2, test_scores[:,1], width, bottom=test_scores[:,0]*0, label="Test F1", color='navajowhite', alpha=0.7)

plt.xticks(x, models)
plt.ylabel("Score")
plt.ylim(0.80, 1.0)
plt.title("Training vs Test Performance  Overfitting Comparison")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(loc='lower right')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Models in desired order
models = ["Overfitted XGBoost", "Dirty Dataset XGBoost", "Best XGBoost"]

# Training scores: [Accuracy, F1]
train_scores = [
    [0.9737, 0.97],   # Overfitted
    [0.9203, 0.92],   # Dirty
    [0.9332, 0.93]    # Best
]

# Test scores: [Accuracy, F1]
test_scores = [
    [0.8775, 0.88],   # Overfitted
    [0.8658, 0.86],   # Dirty
    [0.8736, 0.87]    # Best
]

metrics = ["Accuracy", "F1 Score"]

plt.figure(figsize=(10,6))

# Plot Accuracy
train_acc = [s[0] for s in train_scores]
test_acc = [s[0] for s in test_scores]
plt.plot(models, train_acc, marker='o', label="Train Accuracy", linewidth=2, color='steelblue')
plt.plot(models, test_acc, marker='o', label="Test Accuracy", linewidth=2, color='darkorange')

# Plot F1
train_f1 = [s[1] for s in train_scores]
test_f1 = [s[1] for s in test_scores]
plt.plot(models, train_f1, marker='o', linestyle='--', label="Train F1", linewidth=2, color='skyblue')
plt.plot(models, test_f1, marker='o', linestyle='--', label="Test F1", linewidth=2, color='navajowhite')

# Annotate points
for i, model in enumerate(models):
    plt.annotate(f"{train_acc[i]:.3f}", (model, train_acc[i]), textcoords="offset points", xytext=(0,8), ha='center')
    plt.annotate(f"{test_acc[i]:.3f}", (model, test_acc[i]), textcoords="offset points", xytext=(0,-12), ha='center')
    plt.annotate(f"{train_f1[i]:.2f}", (model, train_f1[i]), textcoords="offset points", xytext=(0,12), ha='center')
    plt.annotate(f"{test_f1[i]:.2f}", (model, test_f1[i]), textcoords="offset points", xytext=(0,-20), ha='center')

plt.ylabel("Score")
plt.ylim(0.80, 1.0)
plt.title("Training vs Test Performance Across XGBoost Variants")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
